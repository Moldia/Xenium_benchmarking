In this notebook we plot some gene we identified to be SVFs to visually inpect wether the scores make sense.

```python
import numpy as np
import scanpy as sc
from scipy.sparse import csr_matrix
import sinfonia
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import scanpy as sc
import os
```

```python
maindir='../../data/unprocessed_adata/'
output_dir='../../data/unprocessed_adata_nuclei/'
files=os.listdir(maindir)
f='h_breast_2.h5ad'
```

```python
adata=sc.read(output_dir+f)
        
```

```python
adata.obsm['spatial']=np.array(adata.obs.loc[:,['x_centroid','y_centroid']])
```

```python
sc.set_figure_params(scanpy=True, dpi=150, dpi_save=150, frameon=True, vector_friendly=True, fontsize=14, figsize=(5,5), color_map=None, format='pdf', facecolor=None, transparent=False, ipython_format='png2x')
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
sc.pl.spatial(adata,spot_size=15,color='GATA3',cmap='PuRd',vmax=20,vmin=-1,show=False)
plt.savefig('../../figures/7.spatial_architecture/GATA3_spatial_hbreast2.pdf')
```
