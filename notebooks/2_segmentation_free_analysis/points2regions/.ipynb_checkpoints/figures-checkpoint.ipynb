```python
adata = ad.read_h5ad(rf"adata_multisection_nuclei_3rep_withHM_withclusters.h5ad")
class_color_palette = dict(zip(np.unique(adata.obs['Class']), adata.uns['Class_colors']))
class_color_palette["Background"] = "#000000" 
```

```python
# Create a pivot table to count occurrences of points2regions per class
pivot_table = df.pivot_table(index='p2r_name', columns='Class', aggfunc='size', fill_value=0)
# pivot_table = pivot_table.fillna(0)
normalized_pivot_table = pivot_table.div(pivot_table.sum(axis=1), axis=0)
#normalized_pivot_table = normalized_pivot_table.fillna(0)
# Calculate the linkage matrix for hierarchical clustering
linkage_matrix = hierarchy.linkage(normalized_pivot_table.values, method='average', metric='euclidean')

# Get the sorted cluster order based on class distribution
sorted_indices = hierarchy.leaves_list(linkage_matrix)

# Sort the pivot table and class index based on the sorted cluster order
sorted_pivot_table = normalized_pivot_table.iloc[sorted_indices]
val = sorted_pivot_table.values - 0.1
max_indices = np.argmax(val, axis=0)
max_indices[val.max(axis=0) < 0] = 1000

# Sort the DataFrame columns based on max_indices
sorted_indices = np.argsort(max_indices)
sorted_pivot_table = sorted_pivot_table[normalized_pivot_table.columns[sorted_indices]]

row_colors = [color_palette.get(point, 'Gray') for point in sorted_pivot_table.index]
```

```python
# Create a clustered heatmap
plt.figure(dpi=160)
g = sns.clustermap(
    sorted_pivot_table, row_cluster=True, col_cluster=False, method='average', metric='euclidean', cmap='viridis', cbar_pos=None, figsize=(13,22),
    row_colors=row_colors, col_colors=sorted_pivot_table.columns.map(class_color_palette), yticklabels=1)

# Add crosses on top of heatmap for the max values of each row, if they contain nuclei in their label
for i, (row_label, row) in enumerate(sorted_pivot_table.iterrows()):
    if "nuclei" in row_label:
        max_index = row.argmax()
        max_value = row.max()
        plt.scatter([max_index + 0.5], [i + 0.5], marker='o', s=50, linewidth=1, facecolors='none', edgecolors='k')

# Reduce y axis tick labels font size
plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)  # ytick rotate
plt.yticks(fontsize=8)
plt.title('Matrix of Distribution of points2regions per Class')
plt.xlabel('Class')
plt.ylabel('points2regions')
# force all y labels to be visible with set_major_locator

plt.tight_layout()
plt.show()
# Save as pdf:
g.savefig("figures/heatmap_p2r_per_class.pdf", dpi=300)
```

```python
selected_classes = ["Oligodendrocytes","Astrocytes"]
for selected_class in selected_classes:
    filtered_df = df[df['closer_clust'] == selected_class]
    # Calculate the median distance for each p2r_name
    median_distances = filtered_df.groupby('p2r_name')['distance_nuc_border'].median().reset_index()

    # Sort the DataFrame by median distance
    sorted_df = filtered_df.merge(median_distances, on='p2r_name', suffixes=('', '_median'))
    sorted_df = sorted_df.sort_values(by='distance_nuc_border_median')

    expression=pd.crosstab(sorted_df['feature_name'],sorted_df['p2r_name'])
    # sort expression columns in same order as sorted_df["p2r_name"].unique()
    expression=expression[sorted_df["p2r_name"].unique()]
    
    expressionfilt=expression.loc[np.max(expression,axis=1)>200,:]
    expressionfilt=expressionfilt.div(expressionfilt.sum(axis=0),axis=1)
    expressionfilt.shape

    expressionfiltnorm=expressionfilt.div(expressionfilt.sum(axis=1),axis=0)
    nbgenes = 7
    # extract top 20 genes for each column
    topgenes=pd.DataFrame(index=expressionfiltnorm.columns,columns=range(nbgenes))
    for i in range(expressionfiltnorm.shape[1]):
        topgenes.iloc[i,:]=expressionfiltnorm.iloc[:,i].sort_values(ascending=False).index[:nbgenes]

    # Make a subplot for each of the p2r_name, with the top nbgenes genes for each p2r_name
    # with x axis being each gene and y axis being the normalized expression value
    nb_plots = expressionfiltnorm.columns.shape[0]
    # use ceil(sqrt) as the number of cols
    nb_cols = nb_plots #int(np.floor(np.sqrt(nb_plots)))
    # compute the number of rows using the number of cols
    nb_rows = 1 #int(np.ceil(nb_plots / nb_cols))

    fig,ax=plt.subplots(nb_rows,nb_cols,figsize=(1.66*len(expressionfiltnorm.columns),4),sharey=True)

    for i, cluster_name in enumerate(expressionfiltnorm.columns):
        gene_expressed = expressionfiltnorm[cluster_name].loc[topgenes.iloc[i,:]].sort_values(ascending=False)
        ax[i].plot(gene_expressed)
        ax[i].set_title(expressionfiltnorm.columns[i].replace("Oligodendrocytes","Oligo"))
        ax[i].set_xticks(range(nbgenes))
        ax[i].set_xticklabels(gene_expressed.index,rotation=90)
        ax[i].set_xlim([-0.5,nbgenes-0.5])
    fig.tight_layout()

    # Save as pdf:
    plt.savefig(f"figures/lineplot_topgenes_{selected_class}.pdf", dpi=300)
    plt.show()

    #filter all expressionfiltnorm where the filter_name is in topgenes
    expressionfilttop=expressionfiltnorm.loc[expressionfiltnorm.index.isin(topgenes.values.flatten()),:]

    # Sort expressionfilttop by the max value in each row in descending order
    expressionfilttop_sorted = expressionfilttop.iloc[expressionfilttop.max(axis=1).argsort()]
    expressionfilttop_sorted

    # Find the column with the maximum value for each row
    max_column = expressionfilttop.idxmax(axis=1)

    # Create a new DataFrame by sorting the original DataFrame based on max_column
    expressionfilttop_sorted = expressionfilttop.iloc[max_column.argsort()]
    plt.figure()
    g = sns.clustermap(expressionfilttop_sorted,row_cluster=False, col_cluster=False, figsize=(6,16))
    plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)  # ytick rotate
    plt.yticks(fontsize=8)

    g.fig.subplots_adjust(right=0.6)
    g.ax_cbar.set_position((0.8, .2, .03, .4))

    # Save as pdf:
    plt.savefig(f"figures/heatmap_topgenes_{selected_class}.pdf", dpi=300)
```
