```python
from somde import SomNode
import scanpy as sc
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
adata=sc.read('../../data/processed_adata/adata_msbrain_3rep_withclusters_only_nuclei_deep_annotation_with_colors.h5ad')
```

```python
adata=adata[adata.obs['replicate']=='1']
```

```python
adata.X
```

```python
df=pd.DataFrame(adata.layers['raw'],columns=adata.var.index)
```

```python
corinfo = adata.obs
corinfo["total_count"]=df.sum(0)
X=corinfo[['x_centroid','y_centroid']].values.astype(np.float32)
```

```python
corinfo[['x_centroid','y_centroid']].values.astype(np.float32)
```

```python
som = SomNode(X,20)
```

```python
ndf,ninfo = som.mtx(df.transpose())
```

```python
ndf
```

```python
nres = som.norm()
```

```python

result, SVnum =som.run()
```
