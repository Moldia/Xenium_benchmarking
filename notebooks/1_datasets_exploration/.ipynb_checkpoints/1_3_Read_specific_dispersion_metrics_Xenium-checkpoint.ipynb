# Objective of the notebook

In this notebook, we calculate the distance from each read in the mouse datasets to its centroid and compute differences between samples and genes

#  Read original unprocessed adata
For each of the datasets, we will read the AnnData containing unprocessed counts and the information of decoded reads in each sample. We will extract the reads, together with the distance to the centroid, cell_id and cell type and store it for later use

```python
# HM #1
adata1=sc.read('../../data/unprocessed_adata/realmouse_1.h5ad')
reads_original=adata1.uns['spots']
adata1=adata[adata.obs['replicate']=='hm1']
id_cluster=dict(zip(adata1.obs['cell_id'],adata1.obs['Class']))
reads_original['initial_annotation_expanded']=reads_original['cell_id'].map(id_cluster)
reads_original_1=reads_original.reset_index()
reads_original_1['ind']=reads_original_1.index
reads_assigned_1=dispersion(reads_original_1,adata1)
reads_assigned_1=reads_assigned_1.loc[:,['feature_name','cell_id','distance','initial_annotation_expanded','overlaps_nucleus']]
reads_assigned_1.to_csv('/media/sergio/xenium_b_and_heart/Xenium_benchmarking-main/data/reads_csvs/realmouse_1_reads_distance.csv')
del adata1
```

# Subsample 10% of the data and combine all `read` objects

For all the read info, we will just keep 10% of the reads for further analysis. In addition, we will combine all dataframes into a single one.

```python
np.random.seed(0)
random_indices=np.random.permutation(list(range(reads_assigned_1.shape[0])))[0:np.int(reads_assigned_1.shape[0]/10)]
ra1=reads_assigned_1.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_2.shape[0])))[0:np.int(reads_assigned_2.shape[0]/10)]
ra2=reads_assigned_2.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_3.shape[0])))[0:np.int(reads_assigned_3.shape[0]/10)]
ra3=reads_assigned_3.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_4.shape[0])))[0:np.int(reads_assigned_4.shape[0]/10)]
ra4=reads_assigned_4.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_preview1.shape[0])))[0:np.int(reads_assigned_preview1.shape[0]/10)]
rp1=reads_assigned_preview1.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_preview2.shape[0])))[0:np.int(reads_assigned_preview2.shape[0]/10)]
rp2=reads_assigned_preview2.iloc[random_indices,:]
random_indices=np.random.permutation(list(range(reads_assigned_preview3.shape[0])))[0:np.int(reads_assigned_preview3.shape[0]/10)]
rp3=reads_assigned_preview3.iloc[random_indices,:]
ra1['sample']='home made 1'
ra2['sample']='home made 2'
ra3['sample']='home made 3'
ra4['sample']='home made 4'
rp1['sample']='preview 1'
rp2['sample']='preview 2'
rp3['sample']='preview 3'
distributions=pd.concat([ra1,ra2,ra3,ra4,rp1,rp2,rp3])
distributions.to_csv('/media/sergio/xenium_b_and_heart/Xenium_benchmarking-main/data/reads_csvs/distributions_all_mouse_brain_10perc.csv')
```

# We format the data, apply kolgomorov-Smirnov and plot distributions

We will divide the data in two groups, based on the source of the experiment (preview or home made). Then we will apply Kolgomorov, smirnov test to check if the distirbution is the same

```python
np.random.seed(0)
random_indices=np.random.permutation(list(range(distributions.shape[0])))[0:np.int(distributions.shape[0]/100)]
distributions=distributions.iloc[random_indices,:]
distributions=distributions.loc[distributions['feature_name'].isin(adata.var.index),:]
repdic={'preview 1':'preview', 'preview 2':'preview', 'preview 3':'preview', 'home made 1':'home made', 'home made 2':'home made', 'home made 3':'home made', 'home made 4':'home made'}
distributions['source']=distributions['sample'].map(repdic)
g1=list(distributions.loc[distributions['source']=='preview','distance'])
g2=list(distributions.loc[distributions['source']=='home made','distance'])
```

```python
ks_2samp(g1, g2)
```

```python
gene_by_ct=pd.DataFrame(index=distributions['feature_name'].unique(),columns=distributions['initial_annotation_expanded'].unique())
for gene in tqdm(gene_by_ct.index):
    for ct in gene_by_ct.columns:
        d1=distributions[distributions['feature_name']==gene]
        d2=d1[d1['initial_annotation_expanded']==ct]
        if d2.shape[0]>10:
            gene_by_ct.loc[gene,ct]=np.mean(d2['distance'])
gene_by_ct[gene_by_ct==-1]=np.nan
meangene=pd.DataFrame(np.nanmean(gene_by_ct,axis=1),gene_by_ct.index)
meancelltype=pd.DataFrame(np.nanmean(gene_by_ct,axis=0),gene_by_ct.columns)
gene_sorted=gene_by_ct.loc[meangene.sort_values(by=0).index,meancelltype.sort_values(by=0).index]
```
